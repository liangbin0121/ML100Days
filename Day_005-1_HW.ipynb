```python
import pandas as pd
import numpy as np
```

```python
p1 = (np.random.randint(1, 10,size=10))*100
data = {'國家': ['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'a8', 'a9', 'a10'],
        '人口': [p1[0], p1[1], p1[2], p1[3], p1[4], p1[5], p1[6], p1[7], p1[8], p1[9] ]}
data = pd.DataFrame(data)
print(data)        
```

```python
data['人口'].max()
```

```python
data.iloc[data['人口'].idxmax()]
```
